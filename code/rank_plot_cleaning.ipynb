```python
# insert average consumption across 
# all countries per year as columns into newdf




# these are all consumption sums of each energy type
biofuel_sum = []
coal_sum = []
fossil_fuel_sum = []
gas_sum = []
hydro_sum = []
low_carbon_sum = []
nuclear_sum = []
oil_sum = []
other_renewable_sum = []
renewables_sum = []
solar_sum = []
wind_sum = []

energy_type_dict = {
    0: biofuel_sum,
    1: coal_sum,
    2: fossil_fuel_sum,
    3: gas_sum,
    4: hydro_sum,
    5: low_carbon_sum,
    6: nuclear_sum,
    7: oil_sum,
    8: other_renewable_sum,
    9: renewables_sum,
    10: solar_sum,
    11: wind_sum
}

energy_values_offset = 2
# starts from year 1900
for year in range(len(world_df)):
    curr_row = world_df.iloc[[year]].values.tolist()[0]

    for c_index in range(3, len(curr_row) - 3):
        curr_col_val = curr_row[c_index]
        # if not NaN values
        if not pd.isna(curr_col_val) and c_index >= 3:
            energy_type_sum = energy_type_dict[c_index - 3]
            energy_type_sum += curr_col_val
            energy_type_dict[c_index - 3] = energy_type_sum
        elif pd.isna(curr_col_val):
            continue
        else:
            raise ValueError("error! not True OR FALSE")




```
